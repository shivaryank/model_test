```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('C:/Users/shiv.kumar/Downloads/test.csv')

# Set the style of the plots
sns.set(style="ticks")

# EDA with Seaborn
# Plot the count of attrition
sns.countplot(x='Attrition', data=data)
plt.title('Attrition Count')
plt.show()

# Plot the distribution of age
sns.histplot(x='Age', data=data, kde=True)
plt.title('Age Distribution')
plt.show()

# Plot the attrition by department
sns.countplot(x='Department', hue='Attrition', data=data)
plt.title('Attrition by Department')
plt.show()

# Plot the attrition by education field
sns.countplot(x='EducationField', hue='Attrition', data=data)
plt.title('Attrition by Education Field')
plt.xticks(rotation=45)
plt.show()

# Plot the attrition by job role
plt.figure(figsize=(12, 6))
sns.countplot(x='JobRole', hue='Attrition', data=data)
plt.title('Attrition by Job Role')
plt.xticks(rotation=45)
plt.show()

# Plot the attrition by job satisfaction
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data)
plt.title('Attrition by Job Satisfaction')
plt.show()

# Plot the attrition by gender
sns.countplot(x='Gender', hue='Attrition', data=data)
plt.title('Attrition by Gender')
plt.show()

# Plot the attrition by monthly income
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data)
plt.title('Attrition by Monthly Income')
plt.show()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Set the style of the plots
sns.set(style="ticks")

# Create subplots for better visualization
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
fig.subplots_adjust(hspace=0.4, wspace=0.3)

# Plot 1: Attrition Count
sns.countplot(x='Attrition', data=data, ax=axes[0, 0])
axes[0, 0].set_title('Attrition Count')
axes[0, 0].set_xlabel('Attrition')
axes[0, 0].set_ylabel('Count')

# Plot 2: Age Distribution
sns.histplot(x='Age', data=data, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Age Distribution')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Count')

# Plot 3: Attrition by Department
sns.countplot(x='Department', hue='Attrition', data=data, ax=axes[1, 0])
axes[1, 0].set_title('Attrition by Department')
axes[1, 0].set_xlabel('Department')
axes[1, 0].set_ylabel('Count')
axes[1, 0].legend(title='Attrition', loc='upper right')

# Plot 4: Attrition by Education Field
sns.countplot(x='EducationField', hue='Attrition', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Attrition by Education Field')
axes[1, 1].set_xlabel('Education Field')
axes[1, 1].set_ylabel('Count')
axes[1, 1].legend(title='Attrition', loc='upper right')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

# Plot 5: Attrition by Job Role
sns.countplot(x='JobRole', hue='Attrition', data=data, ax=axes[2, 0])
axes[2, 0].set_title('Attrition by Job Role')
axes[2, 0].set_xlabel('Job Role')
axes[2, 0].set_ylabel('Count')
axes[2, 0].legend(title='Attrition', loc='upper right')
axes[2, 0].set_xticklabels(axes[2, 0].get_xticklabels(), rotation=45, ha='right')

# Plot 6: Attrition by Job Satisfaction
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data, ax=axes[2, 1])
axes[2, 1].set_title('Attrition by Job Satisfaction')
axes[2, 1].set_xlabel('Job Satisfaction')
axes[2, 1].set_ylabel('Count')
axes[2, 1].legend(title='Attrition', loc='upper right')

# Plot 4: Attrition by Education Field
sns.countplot(x='EducationField', hue='Attrition', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Attrition by Education Field')
axes[1, 1].set_xlabel('Education Field')
axes[1, 1].set_ylabel('Count')
axes[1, 1].legend(title='Attrition', loc='upper right')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

# Plot 5: Attrition by Job Role
sns.countplot(x='JobRole', hue='Attrition', data=data, ax=axes[2, 0])
axes[2, 0].set_title('Attrition by Job Role')
axes[2, 0].set_xlabel('Job Role')
axes[2, 0].set_ylabel('Count')
axes[2, 0].legend(title='Attrition', loc='upper right')
axes[2, 0].set_xticklabels(axes[2, 0].get_xticklabels(), rotation=45, ha='right')

# Plot 6: Attrition by Job Satisfaction
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data, ax=axes[2, 1])
axes[2, 1].set_title('Attrition by Job Satisfaction')
axes[2, 1].set_xlabel('Job Satisfaction')
axes[2, 1].set_ylabel('Count')
axes[2, 1].legend(title='Attrition', loc='upper right')

# Remove empty subplot
fig.delaxes(axes[2, 1])

# Adjust layout to accommodate all subplots
fig.tight_layout()

# Plot 7: Correlation Matrix (Most correlated features)
plt.figure(figsize=(10, 8))
correlation = data.corr()
top_features = correlation['Attrition'].sort_values(ascending=False)[:10].index
sns.heatmap(data[top_features].corr(), annot=True, cmap='coolwarm', ax=plt.gca(), fmt=".2f", annot_kws={"fontsize": 8})
plt.title('Top Correlated Features')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)

plt.show()
```
